Check out the Air B&B dataset, https://raw.githubusercontent.com/nurfnick/Data_Viz/main/AB_NYC_2019.csv. Examine how many entries are null. Impute for atleast two columns that are null in an approriate fashion.

```python
import pandas as pa
df = pa.read_csv('https://raw.githubusercontent.com/nurfnick/Data_Viz/main/AB_NYC_2019.csv')
df.head()
```

```python
df.isnull().sum(axis = 0)
```

```python
df.isnull().sum().sum()
```

Total null value is 20141

let's remove the value null value of reviews_per_month and last_review.

```python
df.reviews_per_month.fillna(value=df.reviews_per_month.mean(), inplace=True)
df
```

```python
df['last_review'] = pa.to_datetime(df['last_review'])
```

```python
df.last_review.dtype #checking dtypes
```

```python
df.last_review.fillna(value=df.last_review.mean(), inplace=True)
df
```

```python
df.isnull().sum(axis = 0)
```

As we can see last_review and number_of_review have now no null value
