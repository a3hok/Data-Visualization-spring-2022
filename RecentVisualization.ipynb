The Indian Premier League (IPL) is a professional Twenty20 cricket league in India that takes place in April and May each year and features teams from various Indian cities and states. The BCCI (Board of Control for Cricket in India) established the league in 2008. The IPL is the most popular cricket league in the world, ranking sixth among all sports leagues in terms of average attendance in 2014.   There is little doubt that the IPL has evolved into a win-win situation for all parties involved, including sponsors, broadcasters, and the BCCI. In the long term, everyone makes money. This is one of the reasons why the company concept is so effective.Matches usually start in the late afternoon or early evening so that at least some of games may be played under floodlights at night to maximize the television audience for international broadcasts. Initially, all league matches were played on a home-and-away basis between all teams, but with the projected expansion to ten clubs (split into two groups of five) in 2011, that format altered, resulting in matches between certain teams being reduced to a single meeting. The top four teams compete in three play-off matches, with one losing club granted a second opportunity to advance to the final, in an effort to maximize possible television income.The four teams who ended at the top of the tables compete in a series of knockout games to provide one team that lost its first-round game a second chance to proceed to the championship match.
