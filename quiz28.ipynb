Using the workout dataset, create a scatterplot with as many features as possible. Can you get 5 or six variables represented in one graphic?

```python
import pandas as ad
import matplotlib.pyplot as plt
import seaborn as sbn
df = ad.read_csv('https://raw.githubusercontent.com/nurfnick/Data_Viz/main/Activity_Dataset_V1.csv')
df.head(10)
```

```python
ax = sbn.scatterplot (data = df, 
                        x = 'workout_type', 
                         y = 'distance',
                       hue = 'workout_type',
                      size = 'calories',
                     legend = 'brief')

sbn.move_legend(ax,"upper left",  bbox_to_anchor=(1.25, 1))
```

It is hard to visualize so lets increase the size.

```python
sbn.set(rc={"figure.figsize":(15, 15)}) #size
ax = sbn.scatterplot(data = df, 
                      x = 'distance', 
                      y = 'time',             
                      size = 'distance',
                     hue = 'workout_type',
                     x_bins = "avg_speed")
sbn.move_legend(ax, "upper right", bbox_to_anchor=(1.25,1))
```

```python
sbn.heatmap(df.corr(), annot=True, linewidths=0.5,vmin = -1) #lets check corelation from heatmap.
```
