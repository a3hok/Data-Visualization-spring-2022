Navigate to the wikipedia page on Marvel Cinematic Universe Films. Gather the table on the films in the Infinity series (Hint: class is 'wikitable plainrowheaders'). Fix any issues with the column names. Remove rows that are not movies.

Firstly, I am loading the liberies from the following code.

```python
import pandas as ad
import requests
from bs4 import BeautifulSoup
```

The wikipedia link for my assignment is  https://en.wikipedia.org/wiki/List_of_Marvel_Cinematic_Universe_films from which i will gather tables.

```python
r = requests.get('https://en.wikipedia.org/wiki/List_of_Marvel_Cinematic_Universe_films')
html_parse = r.text
html_soup = BeautifulSoup(html_parse,'lxml')
```

```python
len(html_soup.find_all('table'))
```

There are 7 table in the list.


Now getting tables from the webpage.

```python
marvelTable = html_soup.find_all('table', class_ = 'wikitable plainrowheaders')
dataFrame = ad.read_html(str(marvelTable))[0]
dataFrame
```

I am going to remove the rows over the heading.

```python
dataFrame.columns = dataFrame.columns.droplevel(1)
dataFrame
```

I can see in the above table that the row number 6 and 13 are not movies. so I have to remove those two rows using the drop method from pandas.

```python
dataFrame = dataFrame.drop(dataFrame.index[[6,13]])
dataFrame
```

the row 6 and 13 are removed.



Now, placeing the table in order. Also i am removing the the unnamed: 5_level_0 from the column too.

```python
dataFrame = dataFrame.reset_index(drop=True, inplace=False)
dataFrame

del dataFrame['Unnamed: 5_level_0']
dataFrame
```

Finally I have removed Unwanted rows that are not movies and unwanted column from the above table.
