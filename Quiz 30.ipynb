```python
import pandas as pa
import matplotlib.pyplot as plt
import seaborn as sbn
df = pa.read_csv('https://raw.githubusercontent.com/nurfnick/Data_Viz/main/Activity_Dataset_V1.csv')
df.head(5)
```

```python
ax = df.groupby('workout_type').calories.agg('mean').plot.barh()
ax.set(title = "Average calories by Workout_type")
ax.set(xlabel = 'Calories')
plt.show()
```

```python
sbn.set(rc={"figure.figsize":(15, 15)})
x = sbn.scatterplot(data = df, 
                      x = 'distance', 
                      y = 'time',             
                      size = 'distance',
                     hue = 'workout_type',
                     x_bins = "avg_speed")
```
