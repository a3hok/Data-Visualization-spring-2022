##Computeing the average time for a trip based on starting point.

```python
%%bigquery --project data-visualization-339004
SELECT 
  start_station_name, 
  AVG(duration_minutes) as Average_time_for_trip
FROM `bigquery-public-data.austin_bikeshare.bikeshare_trips`
WHERE start_station_name is not null 
GROUP BY start_station_name
```

Above is shown the average time for trip.

##Computing how many trips start at each starting point.

```python
%%bigquery --project data-visualization-339004
SELECT 
  start_station_name,
  COUNT(*) as total_trips
FROM `bigquery-public-data.austin_bikeshare.bikeshare_trips`
GROUP BY start_station_name
ORDER BY total_trips DESC
```

Above is shown the total trips.
