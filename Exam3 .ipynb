#c)Create a histogram of the ‘Goals’.

```python
df.Goals.plot.hist(bins= 20, title="Histogram of Goals").set_ylabel("No of players")#histrogram of goals
```

We can see most of them have scored goals.

#d) Create a histogram of the ‘Goals’ and ‘Assists’ on the same axis.

```python
df[['Goals','Assists']].plot.hist(title="Histogram of Goals and Assist").set_ylabel("No of players") #histrogram of goals and assist
```

it shows  most goals are made by assists. There are only very few goals made by individual.

# e) Create a barchart for most goals scored by nationality.

```python
df2 = df.groupby(['nationality'])['Goals'].sum() # Barchart for most goals
df2.plot(kind='bar', title="Barchart of Goals by Nationality").set_ylabel("Total Goals")
plt.show()
```

we can see most goal scored is by Argentina followed by brazil. The least goal is made by Costa Rica.

#f) Create a boxplot of ‘Yel’ (yellow cards) by nationality.

```python
df.boxplot(column = 'Yel', by = 'nationality', rot = 90).set_ylabel("No. of Yellow card")
plt.show()
```

We can see colombia got most yellow card

#h)Create stacked barplot of goals and assists of French National Team. (they won this tourney!) Be sure to comment on why it is appropriate to stack goals and assists.

```python
dfbar = df.groupby(['nationality'])['Goals'].sum()
dfbar.plot(kind='bar', stacked="True", title= "Stacked by Nationality")
plt.show()
```

```python
ax = sbn.heatmap(df.corr()) 
ax.set(title = 'Heamap World cup player')
plt.show()
```
