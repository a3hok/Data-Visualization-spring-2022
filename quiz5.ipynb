#Finding out second most popular starting station.

```python
%%bigquery --project data-visualization-339004
SELECT start_station_name, COUNT(*) as second_popular
FROM `bigquery-public-data.austin_bikeshare.bikeshare_trips`
GROUP BY start_station_name
HAVING second_popular!= (SELECT MAX(second_popular)as max_trip
FROM (
SELECT COUNT(*) as second_popular, start_station_name
FROM `bigquery-public-data.austin_bikeshare.bikeshare_trips`
GROUP BY start_station_name
))
ORDER BY second_popular DESC
LIMIT 1
```

 The above is the second most popular starting station

#computing the round trip which started and completed at the same station.

```python
%%bigquery --project data-visualization-339004

SELECT start_station_name, duration_minutes
FROM `bigquery-public-data.austin_bikeshare.bikeshare_trips`
WHERE start_station_name = end_station_name
```

#computing trips lasted over an hour and were a round trip (started and stopped at the same station)

```python
%%bigquery --project data-visualization-339004

SELECT *, COUNTIF(duration_minutes>60) as trip_over_one_hour
FROM (
  SELECT start_station_name, duration_minutes
FROM `bigquery-public-data.austin_bikeshare.bikeshare_trips`
WHERE start_station_name = end_station_name
)
GROUP BY duration_minutes, start_station_name
Order BY trip_over_one_hour DESC
```
