```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as ad
df = ad.read_csv('https://raw.githubusercontent.com/nurfnick/Data_Viz/main/Activity_Dataset_V1.csv')

df.head(10)
```

```python
plt.pie(x=df.groupby('workout_type').workout_type.agg('count'),labels = df.groupby('workout_type').workout_type.agg('count').index, autopct='%.f' )
plt.show()
```

```python
colors = sns.color_palette('bright')
plt.pie(x=df.groupby('workout_type').workout_type.agg('count'),labels = df.groupby('workout_type').workout_type.agg('count').index, autopct='%.f', colors=colors[1:11]) #color change
plt.show()
```
