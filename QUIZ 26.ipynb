Recreate the three boxplots in this lecture using seaborn. Add the jittered points to the final boxplot.

```python
import matplotlib.pyplot as plt
import pandas as ad

df = ad.read_csv('https://raw.githubusercontent.com/nurfnick/Data_Viz/main/Activity_Dataset_V1.csv')
df.head(10)
```

```python
import seaborn as sns #importing seaborn

ax = sns.violinplot(data = df, x = 'workout_type', y = 'calories')
ax.set_xticklabels(ax.get_xticklabels(),rotation = 90) # increasing the rotation from 45 to 90.
ax.set_title('Violin_Plot_of_Calories_Burned')
sns.stripplot(data = df, x = 'workout_type', y = 'calories' , color = 'black') #stripplot 
plt.show()
```

Now I am adding  the points with some 'jittering', called stripplot in seaborn.

```python
sns.violinplot(data = df, y = 'workout_type', x = 'calories', color = 'black')
sns.stripplot(data = df, y = 'workout_type', x = 'calories' , color = 'green')
```

```python
rgb= [256,56,0]
rgbscaled = []
for color in rgb:
  rgbscaled.append(color/256)
rgbscaled
```

```python
sns.violinplot(data = df, y = 'workout_type', x = 'calories', color = rgbscaled)
sns.stripplot(data = df, y = 'workout_type', x = 'calories' , color = 'blue') # adding jittering
```
