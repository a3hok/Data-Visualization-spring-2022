#(b) Remove the column that repeats the indexes and is ‘unnamed’ as a column.

I am using drop method to delet the column.

```python
df1=df.drop('Unnamed: 0', axis =1)  #delet the column
df1.head(10)
```

#(c) Clean the column names to remove the unit declaration, (%), using regular expressions. The column name should not have any trailing spaces after cleaning. You will only receive partial credit for simply renaming columns without using regular expressions.

Removing unwanted trailing space and sign.

```python
df2 = df1.rename(columns = lambda x: re.sub(r'[(.%)]' ,'' ,x)) #removing (%) from columns
df2.columns
```

#(d) Convert 'activity_day' column into a datetime format.

Converting into datetime format.

```python
df2.activity_day = df2.activity_day.astype('datetime64[ns]') #changing into date time dtypes.
df2.dtypes
```

#(e) Impute ‘total steps’ by replacing the ‘NaN’s with an appropriate number of steps. Convert to appropriate datatype.

First I am using fillna method to replace NaNs. After that I am changing into int dtypes.

```python
df2.total_steps.fillna(value=df2.total_steps.min(), inplace=True)
df2['total_steps'] = df2['total_steps'].astype(int)                #changing into 'int' dypes
df2['total_steps'].head(10)
```

#(f) Convert non-empty ‘avg pace’ into a float that is still representative of the information contained in the column. Recall that there are 60 seconds in one minute so 3:30 is equivalent to 3.5 minutes.

First I am changing ave_pace into date time dtypes from object dtypes.

```python
df2['avg_pace'] = ad.to_datetime(df2['avg_pace'])
df2['avg_pace'] = df2['avg_pace'].dt.hour + (df2['avg_pace'].dt.minute)/60
df2['avg_pace'].head(10)
```

#(g) Group data by ‘workout type’ and find the mean, median, count and standard deviation of ‘calories’.

First I have to  groupby and find mean, median, count and standard deviation of calories.

```python
df2.groupby('workout_type').calories.agg(['mean','median','count', 'std']) # finding mean,median,count and std.
```

#(h) Create an indicator column that identifies if the activity achieved 30% or more ‘aerobic’ activity.

Indicator column that identifies if the activity achieved 30% or more ‘aerobic’ activity.

```python
df2['indicator'] = np.where(df2['aerobic'] >= (30), 'True', 'False')
df2.head(10)
```

#(i) Which day of the week (Monday, Tuesday, etc.) and ‘workout type‘ has the maximum of the ‘max_cadence’.

 Calculating the day, workout type that has maximum  of max_cadence.

```python
df3 = df2.loc[df2.max_cadence == df2.max_cadence.max()]
print('Day  = ' + df3.activity_day.dt.day_name().iloc[0]) #day
print('Workout type = ' + df3.workout_type.iloc[0])     # workout type
print('Max cadence = ' + str(df2['max_cadence'].max()))  # Maximum  cadence
```
